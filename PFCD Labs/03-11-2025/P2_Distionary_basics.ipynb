```python
#Q2) WAP to enter Name & Percentage of marks in a dictionary and Display all Information

d = dict()
n = int(input("Enter the Number of Students: "))
for i in range(n):
    name = input("Enter the Name of the Student: ")
    marks = int(input("Enter the Respective Marks: "))
    d[name] = marks

# for name, value in d.items():
#     print(f"{name} has scored {value}")

for x in d:
    print("Name:",x,"& Marks is:",d[x])
```
