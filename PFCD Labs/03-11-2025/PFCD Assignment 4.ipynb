```python
"""1. Write a Python program to create a list of size N and store the random values in it and find the sum
and average."""

import random
n = int(input("Enter the Number of Values you want to See: "))
min_val = 1
max_val = 10
lis = [random.randint(min_val, max_val) for i in range(n)]

print("Generated List is:",lis)
print("Sum of all Elements present in the List is:",sum(lis))
print("Average of Elements present in the List is:",(sum(lis)/n))
```

```python
"""3. Design and develop a menu-driven Python program for the following list operations.
a. Create a list of N integers
b. Display the list elements
c. Insert an element at a specific position
d. Delete an element at a given position
e. Exit"""

def menu_driven_list_operations():
    my_list = []

    while True:
        print("\n--- List Operations Menu ---")
        print("a. Create a list of N integers")
        print("b. Display the list elements")
        print("c. Insert an element at a specific position")
        print("d. Delete an element at a given position")
        print("e. Exit")
        
        choice = input("Enter your choice (a/b/c/d/e): ").lower()

        if choice == 'a':
            N = int(input("Enter the number of integers (N): "))                
            my_list = []
            for i in range(N):
                element = int(input(f"Enter element {i+1}: "))
                my_list.append(element)
            print("List created successfully.")

        elif choice == 'b':
            if not my_list:
                print("The list is currently empty.")
            else:
                print(f"\nCurrent List Elements: {my_list}")

        elif choice == 'c':
            if not my_list:
                print("List is empty. Create a list first (Option 'a').")
                continue
            else:
                position = int(input(f"Enter the position (0 to {len(my_list)}): "))
                element = int(input("Enter the element to insert: "))
                
                if 0 <= position <= len(my_list):
                    my_list.insert(position, element)
                    print(f"'{element}' inserted at position {position}.")
                else:
                    print(f"Invalid position. Must be between 0 and {len(my_list)}.")

        elif choice == 'd':
            if not my_list:
                print("The list is empty. Nothing to delete.")
                continue
            else:
                print("Elements of the List are:", my_list)
                position = int(input(f"Enter the position to delete (0 to {len(my_list)-1}): "))
                
                if 0 <= position < len(my_list):
                    deleted_element = my_list.pop(position)
                    print(f"Element '{deleted_element}' deleted from position {position}.")
                else:
                    print("Invalid position. Index out of range.")

        elif choice == 'e':
            print("Exiting the program. Goodbye!")
            break

        else:
            print("Invalid choice. Please select a, b, c, d, or e.")

menu_driven_list_operations()
```
