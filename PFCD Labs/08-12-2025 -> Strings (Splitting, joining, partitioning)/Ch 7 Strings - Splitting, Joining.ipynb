```python
# Splitting Strings

l = 'A, B, C, D'
print(l.split(','))
print(l.split(',' , 2))
print(l.rsplit(',' , 2))
```

```python
# Joining Strings

','.join(["A", "B", "C", "D"])
','.join(str(i) for i in range(1,11))
```
