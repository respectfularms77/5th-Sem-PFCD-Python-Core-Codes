```python
#View only creates the Mirror of the Parent's ndarray and modifying the parent results in the modification
#CHild's too.....

import numpy as np
n = np.arange(1,6)

v = n.view()
print("Normal:",n,"\tView:",v)
print("\nNormal's ID:",id(n),"\tView's ID:",id(v))
```

```python
#Modifying the Items only on the Parent's ndarray

n[1] *= 10
print("Normal:",n,"\tView:",v)
```

```python
#Applying the Slicing techniques on the Parents ndarray

n2 = n[0:3]
print("Normal:",n,"\nNew One:",n2)
```

```python
n2[1] = 55
print(n2)
```

```python
#Deep Copy is the Independent Copy of the Original Data
import copy
n = np.arange(1,11)
d = n.copy()
print("N:",n, "\tShallow Copy:",d)
print("\nOriginal's ID:", id(n), "\tShallow Copy's ID:",id(d))


#Modifying the Shallow Copy doesn't results in chainging of the Parent file
d[1] = 100
print("\nN:",n, "\tShallow Copy:",d)
```

```python
import numpy as np

# Original array
arr = np.array([1, 2, 3, 4, 5, 6])
print(f"Original array: {arr}")
print(f"Shape of original array: {arr.shape}\n")

# Using reshape()
reshaped_arr = arr.reshape(2, 3)
print(f"Array after reshape: \n{reshaped_arr}")
print(f"Shape of reshaped array: {reshaped_arr.shape}")
print(f"Original array after reshape (unchanged): {arr}\n")

# Using resize() - method
print("Using array.resize():")
arr.resize(3, 2) # Modifies arr in-place
print(f"Original array after in-place resize: \n{arr}")
print(f"Shape of original array after in-place resize: {arr.shape}\n")
```

```python
#Hstack anf Vstack
grade = np.array([[98,97,96],[100,101,140]])
grade1 = np.array([[92,95,94],[90,89,101]])

print("Horizontal Stack:\n",np.hstack((grade,grade1)))
print("\n\nVertical Stack:\n",np.vstack((grade,grade1)))
```
